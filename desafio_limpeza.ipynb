# Conversão dos dados

```python
df.info()
```

```python
def converter_colunas(df):
  df_copia = df.copy()
  for coluna in df_copia.columns:
        if coluna.lower() == 'idade' or coluna.lower() == 'qtd_compras_12m' or coluna.lower() == 'renda_mensal':
          df_copia[coluna] = pd.to_numeric(df_copia[coluna], errors='coerce').astype('Int64')
        elif coluna.lower() == 'score_credito':
            df_copia[coluna] = pd.to_numeric(df_copia[coluna], errors='coerce')
  return df_copia

df_convertido = converter_colunas(df)
df_convertido.info()
```

```python
df_convertido['renda_mensal'] = df_convertido['renda_mensal'].mask(df_convertido['renda_mensal'] < 0, np.nan)
# O método mask() do pandas é usado para substituir valores em um DataFrame onde uma condição é verdadeira
```

# Preenchimento de valores ausentes

```python
df_convertido.isnull().sum()
```

```python
df_convertido['idade'].unique()
```

```python
df_convertido
```

```python
for col in ['idade', 'score_credito', 'qtd_compras_12m']:
    df_convertido.loc[:, col] = df_convertido.loc[:, col].astype(float).fillna(df_convertido.loc[:, col].mean())

df_convertido.loc[:, 'renda_mensal'] = df_convertido.loc[:, 'renda_mensal'].fillna(df_convertido.loc[:, 'renda_mensal'].median())

def fill_with_mode(series):
    mode = series.mode()
    if not mode.empty:
        return series.fillna(mode[0])
    else:
        return series.fillna("Desconhecido")

for col in ['estado_civil', 'uf', 'categoria_produto']:
    df_convertido.loc[:, col] = fill_with_mode(df_convertido.loc[:, col])

df_convertido.loc[:, 'data_cadastro'] = df_convertido.loc[:, 'data_cadastro'].fillna(df_convertido.loc[:, 'data_cadastro'].mode()[0] if not df_convertido.loc[:, 'data_cadastro'].mode().empty else "Desconhecido")
```
